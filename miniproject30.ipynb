/////Basic subplot visualization/////

/////***Line chart****////

```python
import matplotlib.pyplot as plt
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
sales = [150, 200, 180, 220, 260, 300, 280]
plt.figure(figsize=(10,8))
plt.plot(days,sales,color='purple',marker='*',markerfacecolor='b',linestyle='-.',label='weekly sales report')
plt.xlabel('days')
plt.ylabel('sales')
plt.title("sales report")
plt.legend(loc='upper left')

plt.show()
           
```

   ////****scatter plot chart***//

```python

plt.figure(figsize=(10,8))
plt.scatter(days,sales,color="pink")
plt.xlabel('days')
plt.ylabel('sales')
plt.title("sales report")


plt.show()
```
