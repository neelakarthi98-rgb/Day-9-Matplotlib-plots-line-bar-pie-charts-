
/////Sales Data Analysis with Bar and Pie Charts////

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
sales_data={"category":['furniture','Technology','office suplies','Technology','furniture','office suplies'],
            "sales":[15000,20000,25000,9000,11000,5000]}
df=pd.DataFrame(sales_data)
print(df)
```

/////Aggregate Data////

```python
total_sales=df.groupby('category')['sales'].sum()
print(total_sales)
print()
total_sales_count=total_sales.reset_index()
total_sales_count.columns=['category','total_sales']
df1=pd.DataFrame(total_sales_count)
print(df1)
```

////create a bar chart using Aggregate data///

```python
plt.figure(figsize=(5,5))
plt.bar(df1['category'],df1['total_sales'],color='green')
plt.title("total sales per category")
plt.xlabel('category')
plt.ylabel('total sales')

plt.tight_layout()
plt.show()
```

////pie chart//

```python
plt.figure(figsize=(10,5))
plt.pie(df1['total_sales'],labels=df1['category'],autopct='%1.1f%%',startangle=90,colors=['skyblue','orange','green'])
plt.title("Total sales per category")
plt.show()
```

/////subplot///

```python

plt.figure( figsize=(8,8))
plt.subplot(1,2,1)
colors=('green','blue','yellow')
plt.bar(df1['category'],df1['total_sales'],color=colors)
plt.title("Total sales per product")
print()
plt.subplots_adjust(wspace=0.4, hspace=0.6)

plt.subplot(1,2,2)
plt.pie(df1['total_sales'],labels=df1['category'],autopct='%1.1f%%',startangle=90,colors=['skyblue','orange','green'])
plt.title("Total sales per product")
plt.show()
```
