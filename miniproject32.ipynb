////////Student Test Score Analysis////

//Visualize test score distributors for students from different schools and identify top performing groups///

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
score_data={'studentname':['Arul','charlie','bob','thomas','Arul','charlie','bob','thomas','charlie','bob','Arul','thomas'],
"score":[85,60,75,50,55,45,95,70,90,65,90,91],
"school_name":['vvd','dvd','msd','svd','vvd','dvd','msd','svd','dvd','msd','vvd','svd']}
df1=pd.DataFrame(score_data)


```

///////overall score distribution using hisplot

```python
plt.Figure(figsize=(10,8))
sns.histplot(data=df1,x='score',bins=10,kde=True)
plt.title("overall score for each student")
plt.xlabel("student score")
plt.ylabel("studentname")


plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

///**compare scores by school using boxplot**//

```python
plt.figure(figsize=(5,5))
sns.boxplot(data=df1,x='school_name',y='score',hue="school_name",palette='Set2')
plt.xlabel("school name")
plt.ylabel("scores")
plt.title("scores by school")
plt.show()
```
