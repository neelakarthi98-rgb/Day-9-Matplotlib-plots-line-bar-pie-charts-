///////Titanic Survival Analysis/////

///****Use a real world dataset to visualize survival raters based on categorial features**//

```python
import seaborn as sns
import matplotlib.pyplot as plt
data=sns.load_dataset('titanic')


```

///**Survival Bar Plot**//

```python
colors=('skyblue','green','yellow')
plt.Figure(figsize=(5,5))

sns.barplot(data=data,x='class',y='survived',hue='sex',palette=colors)
plt.xlabel('class')
plt.ylabel('Survived')

plt.title("Average Survival Rate")
plt.tight_layout()
plt.show()
```

////create violinplot///

```python
plt.figure(figsize=(5,5))
colors={0:'yellow',1:'blue'}
sns.violinplot(data=data,x='sex',y='age',hue='survived',palette=colors)
plt.xlabel("Sex")
plt.ylabel('Age')
plt.title("Distribution of Survivor and non-survivors")
plt.show()
```
